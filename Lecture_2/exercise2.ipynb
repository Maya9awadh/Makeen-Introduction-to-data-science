```python
#imported library
from random import randint
```

```python
#create a list of integers
rand_int=[]
for i in range(1,1001):
    num=randint(1,1000)
    rand_int.append(num)
```

```python
#calculate mean 
mean = sum(rand_int) / len(rand_int)
mean
```

```python
differences = [(value - mean)**2 for value in rand_int]#x-mean\
sum_diff=sum(differences) #sum of x-mean
stand_d=(sum_diff / (len(rand_int) - 1)) ** 0.5 #standard deviation
stand_d
```

```python
#calculate z-score
z_scores = [(value - mean) / stand_d for value in rand_int] #calculate z-score
z_scores
```

```python
outliers=[]
for score in z_scores:
    if score>2:
        outliers.append(score)
print(outliers)
```
